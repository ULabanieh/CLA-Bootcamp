```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("abcsds/pokemon")

print("Path to dataset files:", path)
```

```python
import os
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
os.listdir(path)
```

```python
df = pd.read_csv(os.path.join(path, "Pokemon.csv"))
df.head()
```

```python
plt.figure()
sns.scatterplot(df, x="Attack", y="Defense")
plt.title("Attack-Defense Scatterplot")
plt.savefig("attack_defense_scatterplot.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
plt.figure()
sns.histplot(df,x="Speed")
plt.title("Speed Distribution")
plt.savefig("speed_distribution.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
plt.figure()
sns.lineplot(df, x="Generation", y = "Attack", errorbar=None)
plt.title("Attack Evolution")
plt.savefig("attack_evolution.png", dpi=300, bbox_inches='tight')
plt.show()

plt.figure()
sns.lineplot(data=df, x="Generation", y="Defense", color="#C4432B", errorbar=None)
plt.title("Defense Evolution")
plt.savefig("defense_evolution.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
plt.figure()
plt.title("Number of Legendary vs Non-Legendary Pokemon")
sns.countplot(x="Legendary", data=df)
plt.xlabel("Legendary Status")
plt.ylabel("Pokemon Count")
plt.savefig("legendary_pokemon_count.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
corr = df[["HP", "Attack", "Defense", "Sp. Atk", "Sp. Def", "Speed"]].corr()

plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, cmap="Reds")
plt.title("Pokemon Stats Heatmap")
plt.savefig("pokemon_stats_heatmap.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
plt.figure()
sns.pairplot(df[["Total", "HP", "Attack", "Defense", "Sp. Atk", "Sp. Def", "Speed"]])
plt.title("Pairplot of Numeric Features")
plt.savefig("pairplot.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.violinplot(x="Type 1", y="Attack", data=df)
plt.xticks(rotation=45)
plt.title("Attack Violinplot by Pokemon Type")
plt.savefig("violinplot.png", dpi=300, bbox_inches='tight')
plt.show()
```
