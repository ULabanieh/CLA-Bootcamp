## Data visualization
The function plot_two_classes is a Python function that creates a scatter plot to visualize two classes of data based on the input features X1 and X2.

```python
def plot_two_classes(X_df, y_df):
    plt.scatter(X_df[y_df == 0]['X1'], X_df[y_df == 0]['X2'], c='blue', label='Class 0', alpha=0.7)
    plt.scatter(X_df[y_df == 1]['X1'], X_df[y_df == 1]['X2'], c='red', label='Class 1', alpha=0.7)
    plt.xlabel('X1')
    plt.ylabel('X2')
    plt.title('Scatter Plot of Classes')
    plt.legend()
    plt.show()
```

## 1 - Oversampling the Minority Class:

In this section, we will use the SMOTE technique that allows to create synthetic samples of the minority class in a dataset to balance the class distribution. It is particularly useful when you have a classification problem with an imbalanced dataset, meaning one class significantly outnumbers the other(s).

To do:

- use the same csv file and drop the Y column from dataFrame to keep only the X1 and x2 after that save the Y in new variable 'y' (target variable)
- Create the smote Instance.
- Resample your dataset with a new oversampled dataset using the fit_resample function of the smote technique.
- Visualize the result in plot using the function plot_two_classes.



```python
df = pd.read_csv('Datasets/imbalanced.csv')# read the csv file again
```

```python
X = df.drop('Y', axis= 1) # feature matrix
y = df['Y'] # target variable
```

```python
from imblearn.over_sampling import SMOTE

df = pd.read_csv('Datasets/imbalanced.csv') # read the csv file again
X = df.drop('Y', axis= 1 ) # feature matrix
y = df['Y'] # target variable
# create the instance smote here
smote = SMOTE()
X_resampled, y_resampled = smote.fit_resample(X, y)

# add the plot here
plot_two_classes(X_resampled, y_resampled)
```

## 2 - Undersampling the Majority Class:

Now we need to read again the csv file then do the following steps to perform a SMOTE(  Undersampling Technique).
To do :  
- same as before remove the Y column and keep just X1 and X2 .
- Create instance of EditedNearestNeighbours.
- Resample your dataset with a new undersampling  using the fit_resample function of the EditedNearestNeighbours technique.
- To do comparision between the new balanced dataset and the original dataset visualize both the new dataset and the original dataset from the csv file using the function plot_two_classes.

```python
from imblearn.under_sampling import EditedNearestNeighbours
import pandas as pd

df = pd.read_csv('Datasets/imbalanced.csv')
X = df.drop('Y', axis=1) # feature matrix
y = df['Y'] # target variable
enn = EditedNearestNeighbours()
X_resampled, y_resampled = enn.fit_resample(X, y)

plot_two_classes(X, y)
plot_two_classes(X_resampled, y_resampled)
```

## 3 - Combination of Oversampling and Undersampling:

Now we need to read again the csv file then do the following steps to perform a combination of SMOTE( Over-sampling and Undersampling Technique).
To do :  
- same as before remove the Y column and keep just X1 and X2 .
- Create instance of SMOTETomek.
- Resample your dataset with a new combined oversampling and undersampling dataset using the fit_resample function of the SMOTETomek technique.
- To do comparision between the new balanced dataset and the original dataset visualize both the new dataset and the original dataset from the csv file using the function plot_two_classes.

```python
from imblearn.combine import SMOTETomek
import pandas as pd

df = pd.read_csv('Datasets/imbalanced.csv')
X = df.drop('Y', axis=1) # feature matrix
y = df['Y'] # target variable
smt = SMOTETomek()
X_resampled, y_resampled = smt.fit_resample(X, y)

plot_two_classes(X, y)
plot_two_classes(X_resampled, y_resampled)
```
