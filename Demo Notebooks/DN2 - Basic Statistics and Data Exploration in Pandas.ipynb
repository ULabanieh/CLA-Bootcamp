# Understanding Data Types in Data Science: A Guide with Pandas

Data science involves the analysis and prediction based on datasets, where understanding the data structure is crucial. Data can be broadly classified into two types: numerical and categorical.

- **Numerical Data :**
Numerical data comprises values that can be measured quantitatively, including integers, floats, and continuous values. Examples include age, weight, height, and temperature.

- **Categorical Data :**
Categorical data represents discrete categories or labels, such as gender, nationality, or color. There are two subtypes:



1. Nominal Data: No inherent order (e.g., eye color).
2. Ordinal Data: Has a natural order (e.g., clothing size).

## Pandas Data Types

# Basics of Statistics in Pandas:

**Definition:** Statistics is the study of data, involving the collection, analysis, and interpretation of information. In data science, it helps make sense of large datasets and draw meaningful conclusions.

**Importance:** Understanding statistics is crucial for data scientists to make informed decisions and effectively communicate their findings.



## Descriptive Statistics:

**Definition:** Descriptive statistics summarizes and describes the features of a dataset.

## Common Measures:

- **Measures of Central Tendency:** Describing the typical value of a dataset.
- **Measures of Variability:** Describing how spread out the values are.
- **Measures of Shape:** Measures of shape are statistical metrics used to describe the overall shape of a dataset.
<br><br>

## Measures of Central Tendency:

#### a. Mean:

**Definition:** The average value of a dataset.

**Formula:** Mean = Sum of values / Number of values

**Example:** For the dataset [2, 4, 6], the mean is (2 + 4 + 6) / 3 = 4.
<br><br><br>

#### b. Median:

**Definition:** The middle value when the dataset is ordered.

**Example:** For [2, 4, 6, 8], the median is 5 (average of 4 and 6).
<br><br><br>

#### c. Mode:

**Definition:** The most common value in a dataset.

**Example:** In [2, 2, 4, 6, 6], the mode is 2 and 6.
<br><br><br>


```python
import pandas as pd

df = pd.read_csv("iris.csv")
df
```

```python
 #describe
 df.describe()
```

```python
#sepal.length
mean = df['sepal.length'].mean()
median = df['sepal.length'].median()
mode = df['sepal.length'].mode()
print(mean, median, mode)
```

## Measures of Variability:

#### a. Range:

**Definition:** The difference between the maximum and minimum values.

**Example:** For [3, 7, 1, 9], the range is 9 - 1 = 8.
<br><br><br>

#### b. Variance:

**Definition:** A measure of how spread out values are from the mean.

**Formula:** Variance = Sum of squared differences from mean / Number of values

**Example:** For [1, 2, 3], the mean is 2, and variance is ((1-2)^2 + (2-2)^2 + (3-2)^2) / 3 = 2/3.
<br><br><br>
#### c. Standard Deviation:

**Definition:** The square root of the variance, indicating spread from the mean.

**Example:** If the variance is 2/3, the standard deviation is sqrt(2/3).

```python
#sepal.width
range = df['sepal.width'].max() - df['sepal.width'].min()
variance = df['sepal.width'].var()
std_deviation = df['sepal.width'].std()

print(range, variance, std_deviation)
```

## Measures of Shape:

**Definition:** Measures of shape are statistical metrics used to describe the overall shape of a dataset. Skewness and kurtosis are two common measures that provide insights into the asymmetry and "peakedness" of the distribution.

## 1. Skewness:

**Definition:** Skewness measures the asymmetry of a dataset's distribution.

- **Positive Skewness:** The tail of the distribution is longer on the right side, indicating that the right side is stretched or has outliers.
  - **Example:** Consider income data where most people earn average salaries, but a few individuals have exceptionally high incomes.

- **Negative Skewness:** The tail of the distribution is longer on the left side, indicating that the left side is stretched or has outliers.
  - **Example:** Examining the distribution of expenses, where most people spend moderately, but a few have very low expenses.

## 2. Kurtosis:

**Definition:** Kurtosis measures the "peakedness" of a dataset's distribution.

- **High Kurtosis:** Indicates a sharp peak in the distribution, suggesting that data points are concentrated around the mean with heavy tails.
  - **Example:** A dataset representing the heights of professional basketball players might exhibit high kurtosis, as most players are of similar heights.

- **Low Kurtosis:** Indicates a flat peak, suggesting that data points are more spread out from the mean with lighter tails.
  - **Example:** Examining the distribution of heights in a random sample of people might show low kurtosis, as heights vary more widely.

Understanding skewness and kurtosis provides additional information about the characteristics of a dataset beyond measures of central tendency and variability. These measures are particularly useful in identifying patterns and outliers in the distribution of data.


Kurtosis is sometimes reported as “excess kurtosis.” Excess kurtosis is determined by subtracting 3 from the kurtosis. This makes the normal distribution kurtosis equal 0.

```python
skewness = df["sepal.width"].skew()
kurtosis = df["sepal.width"].kurtosis()
print(skewness, kurtosis)
```

* Skewness = 0: Then normally distributed.
* Skewness > 0: Then more weight in the left tail of the distribution.
* Skewness < 0: Then more weight in the right tail of the distribution.
---
* kurtosis for normal distribution is equal to 3.
* For a distribution having kurtosis < 3: It is called playkurtic.
* For a distribution having kurtosis > 3, It is called leptokurtic and it signifies that it tries to produce more outliers rather than the normal distribution.

## Inferential Statistics in Data Analytics:

**Definition:** Inferential statistics is a branch of statistics that involves making predictions or inferences about a population based on a sample of data. The goal is to estimate population parameters using sample data.

**Importance in Data Analytics:**

In data analytics, we often deal with large datasets. Instead of analyzing every single data point in a population, we can use inferential statistics to draw conclusions about the population based on a representative sample. This makes the analysis more feasible, cost-effective, and efficient.

## Confidence Intervals:

**Definition:** Confidence intervals are used to estimate the range of values in which the true population parameter is likely to lie. The confidence interval is calculated based on the sample mean and the standard deviation of the sample.

**Purpose:**
- Provide a range of values for the population parameter.
- Reflect the uncertainty inherent in statistical estimation.
- Allow data analysts to express the precision of their estimates.

**Example:**
Consider a survey estimating the average income of a population. A 95% confidence interval might be [\$40,000, \$50,000], indicating we are 95% confident that the true average income falls within this range.



```python
import pandas as pd
import scipy.stats as stats

df = pd.read_csv("iris.csv")
sample_mean = df["sepal.width"].mean()
sample_std = df["sepal.width"].std()

conf_int = stats.t.interval(0.95, len(df)-1, loc=sample_mean, scale=sample_std)

print("conf_int :",conf_int)
```

## Hypothesis Testing:

**Definition:** Hypothesis testing is used to test a hypothesis about a population parameter based on a sample of data. The hypothesis is typically in the form of a null hypothesis, which assumes that there is no significant difference between the sample and the population. The alternative hypothesis assumes that there is a significant difference.

**Purpose:**
- Evaluate claims or hypotheses about population parameters.
- Provide a formal and systematic way to make decisions based on data.

**Example:**
Suppose we want to test if a new website design improves user engagement. The null hypothesis could be that there is no significant difference, while the alternative hypothesis suggests there is a significant improvement. By analyzing user engagement data, we can decide whether to accept or reject the null hypothesis.

**Why We Do Them in Data Analytics:**

1. **Efficient Decision-Making:** Inferential statistics allow us to make informed decisions about populations based on limited sample data, saving time and resources.

2. **Quantifying Uncertainty:** Confidence intervals quantify the uncertainty associated with our estimates, providing a more nuanced understanding of the population parameters.

3. **Scientific Rigor:** Hypothesis testing adds a layer of scientific rigor to data analytics, helping ensure that findings are not just due to random chance.

```python
import pandas as pd
import scipy.stats as stats

# Assuming "data.csv" contains your data and "column" is the column you're interested in
df = pd.read_csv("iris.csv")
test_value = 3

# Perform one-sample t-test
test_result = stats.ttest_1samp(df["sepal.width"], test_value)

# Print the result
print("T-statistic:", test_result.statistic)
print("P-value:", test_result.pvalue)
```

### T-statistic:

- The t-statistic measures how far the sample mean is from the hypothesized population mean (test_value) in terms of standard errors.
- A higher absolute t-statistic indicates a greater difference between the sample mean and the hypothesized mean.

### P-value:

- The p-value is a measure of evidence against a null hypothesis. It tells you the probability of observing the given t-statistic (or more extreme) if the null hypothesis is true.
- A smaller p-value (typically below a chosen significance level, e.g., 0.05) suggests stronger evidence against the null hypothesis.

### Interpretation:

- If the p-value is less than your chosen significance level (e.g., 0.05), you may reject the null hypothesis.
- If the p-value is greater than the significance level, you may not reject the null hypothesis.

### Example Interpretation:

- If `test_result.pvalue` is 0.02 (less than 0.05), you might say: "There is statistically significant evidence to reject the null hypothesis that the population mean is 50 in favor of the alternative hypothesis."
- If `test_result.pvalue` is 0.15 (greater than 0.05), you might say: "There is not enough evidence to reject the null hypothesis that the population mean is 50."


In summary, inferential statistics, including confidence intervals and hypothesis testing, are crucial tools in data analytics. They enable analysts to make meaningful predictions about populations and draw reliable conclusions from sample data.
