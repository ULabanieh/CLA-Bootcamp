# Data Visualization

### What is Data Visualization

Data visualization is the graphical representation of data and information. It is an essential tool for data scientists, analysts, and decision-makers to help them make sense of complex data sets. The power of data visualization is that it helps us see patterns, relationships, and trends that might not be apparent from raw data or tables.

### Significance of Data Visualization

The human brain is wired to process visual information faster than any other type of information. We process visuals 60,000 times faster than text, and 90% of the information transmitted to our brain is visual. This is why data visualization is so important; it allows us to process and analyze large amounts of data quickly and efficiently.

### Data Visualization and Statistical Conclusions

Data visualization is not just a way to present data; it also plays a crucial role in statistical analysis. When we analyze data, we often use statistical tests to determine if there are significant differences or relationships between variables. Data visualization helps us visualize the results of these statistical tests, making it easier to draw conclusions from the data.

## Scatter Plots

A scatter plot identifies a possible relationship between changes observed in two different sets of variables. It provides a visual and statistical means to test the strength of a relationship between two variables.

### Plotting a Scatter Plot in Python

```python
# import seaborn
import seaborn as sns

# simple scatter plot
sns.scatterplot(x='Price', y='SqFt', data=housing_df)
```

### When to use a Scatter Plot

- When you have paired numerical data
- When your dependent variable may have multiple values for each value of your independent variable
- When trying to determine whether the two variables are related

## Histogram

A histogram is used to visualize the distribution of a continuous variable. The x-axis represents the range of values of the variable, and the y-axis represents the frequency or proportion of observations within each range. Histograms are useful for identifying the shape of the distribution, as well as for identifying outliers and skewness in the data.

### Plotting a Histogram in Python

```python
column = "Bedrooms"

sns.histplot(data=housing_df, x=column)
```

### Why use Histograms

One of the main advantages of using a histogram is that it can help you analyze the distribution and variation of a continuous variable, which can reveal important insights about the process or the population.

## Line Plot

A line plot is used to show trends over time or to display relationships between two variables.

### Plotting Line Plot in Python

```python
sns.lineplot(data=housing_df, x="Price", y="Bedrooms")
# housing_df["Price"].to_numpy()
```

### When to use a Line Plot

1. Trend Visualization:
    1. Use a line plot when you want to show the overall trend or pattern in your data over a continuous variable, such as time.

2. Comparing Multiple Trends:
    1. Employ a line plot when you need to compare multiple trends on the same graph, making it easy to see relative changes.
3. Sequential Data Display:
    1. Opt for a line plot when you want to display the progression of values in a sequential order, highlighting the flow of data.

4. Correlation Assessment:
    1. Use a line plot to assess the correlation between two variables, observing how changes in one variable relate to changes in another.

5. Frequency Distribution:
    1. Choose a line plot when representing the distribution of values, particularly when the x-axis represents categories or bins.

6. Monitoring Progress:
    1. Utilize a line plot when you need to monitor the progress of a variable over time or against specific milestones.

## Heat Maps

A heatmap is used to visualize the relationship between two categorical variables. The colors in the heatmap represent the frequency or proportion of the combination of the two variables. Heatmaps are useful for identifying patterns and trends in the data, as well as for identifying areas with high or low frequency or proportion.

### HeatMaps in Python
