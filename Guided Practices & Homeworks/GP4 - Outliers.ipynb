# Outliers

Outliers are data points that significantly deviate from the rest of the observations in a dataset. They can result from measurement errors, data entry mistakes, or rare events. Identifying and managing outliers is crucial in data analysis to prevent skewed results and erroneous conclusions.

## Reasons for Outliers

- **Measurement Error:** Errors in the measurement process can lead to extreme values that do not accurately represent the observed phenomenon.

- **Data Entry Errors:** Mistakes in entering data can introduce values significantly different from other observations.

- **Natural Variation:** Outliers may represent real but rare events that deviate from the typical pattern in the data.

- **Extreme Values:** Outliers can arise from extreme values inherent in the underlying distribution, especially in heavy-tailed distributions.

Regardless of the cause, outliers can distort data exploration and analysis, impacting the reliability of results.

## Methods of Identifying Outliers

1. **Visual Inspection:** Utilizing visualization techniques such as boxplots, scatterplots, and histograms allows the identification of outliers as observations deviating significantly from the overall pattern.

   *Example:* In a scatterplot, an outlier can be visually spotted as a data point deviating markedly from the others.

2. **Descriptive Statistics:** Mean, median, and standard deviation can be employed to identify outliers. Observations falling outside a certain range or threshold are flagged as outliers.

   *Example:* Identifying outliers by calculating the mean and standard deviation, then flagging observations beyond two standard deviations from the mean.

3. **Machine Learning Models:** Algorithms like isolation forests, one-class SVMs, and local outlier factor (LOF) can be applied to identify observations that deviate significantly from the majority.

   *Example:* Using the LOF algorithm to identify outliers in a dataset and flagging observations with a negative outlier factor below a threshold.


# outliers

In this practice, we will learn how to identify outliers by using a simple example. We'll start by generating a dataset that follows a normal distribution.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate random data
data = np.random.normal(0, 1, 100)

# Add an outlier
data[0] = 10
```

# Methods of Identifying Outliers:

## 1 - Visual inspection:

Visualize the generated data in scatter plot of matplotlib

```python
# Create a scatterplot
plt.scatter(range(len(data)), data)
plt.show()
```

## 2 - Descriptive statistics:

- Compute the mean and std than compare each data point to mean-2 * std and mean +2 * std :

  --> if x it's less than mean -2 *std or greater than mean + 2 * std than an outlier us identified.

```python
# compute the mean and std
mean = data.mean()
std = data.std()
outliers=[]
# Identify outliers

for x in data:
   if (x < mean - 2 * std) or (x > mean + 2 * std):
      outliers.append(x)
print(outliers)
```

# Methods of Dealing with Outliers:

## 1 - Removing outliers:

keep only data that have values less than 6 and plot it in scatter

```python
data_without_outlier = data[data < 6]

# Create a scatterplot
plt.scatter(range(len(data_without_outlier)), data_without_outlier)
plt.show()
```

## 2 - Transforming the data:

a log transformation can be used to handle data with extreme values.

```python
# Apply log transformation
data_transformed = np.log(data - np.min(data) + 1)

#plot the result
plt.scatter(range(len(data)), data)
plt.scatter(range(len(data_transformed)),data_transformed)
plt.show()
```
