# Dealing with Missing Values

## The Problem of Missing Values

A missing value in a dataset is a value that is not recorded. Various reasons, such as human error or equipment failure, can lead to missing values. These gaps can impact data analysis by:

- **Biasing Statistical Estimates:** Missing values can introduce bias into statistical estimates, making it challenging to generalize results.
- **Reducing Sample Size:** The presence of missing values decreases the sample size, affecting the statistical power of the analysis.
- **Creating Inconsistencies:** Missing values can lead to inconsistencies in the data, causing errors in analysis and interpretation.

Addressing missing values is essential for accurate data analysis.

## Methods to Deal with Missing Values

### 1 - Dropping Missing Values

Dropping missing values is a straightforward approach involving the removal of rows or columns containing missing values. This method ensures a complete dataset but may reduce the sample size and lead to biased estimates if missing values are not randomly distributed.

### 2 - Imputing Missing Values

Imputing missing values involves estimating them based on other variables in the dataset. Mean, median, or mode imputation are common methods. Mean imputation replaces missing values with the mean of the variable, preserving the sample size.

### 3 - Using a Model to Impute Missing Values

Advanced imputation involves building a model to predict missing values based on other variables. This method, employing machine learning algorithms like linear regression, is effective but requires expertise and can be time-consuming.

#### 1. Mean Imputation

Mean imputation replaces missing values with the mean of non-missing values in the same column. It's simple but may distort the variable's distribution if missing values are not random.


```python
# Assuming 'column_name' is the column where you want to perform mean imputation
mean_value = df['Age'].mean()
print ("mean age: ",df['Age'].mean())

# Replace missing values with the mean value in the specified column
df['Age'].fillna(mean_value, inplace=True)

# printing mean age of the column
print("Mean Age: ",df['Age'].mean())

# Display the DataFrame after mean imputation
df
```

#### 2 - Median Imputation

Similar to mean imputation, median imputation uses the median value instead. It's less sensitive to outliers and can be more suitable for variables with extreme values.

```python
# Assuming 'column_name' is the column where you want to perform mean imputation
median_value = df['Age'].median()

# Replace missing values with the mean value in the specified column
df['Age'].fillna(median_value, inplace=True)

# printing our the median of the column
print("Median Value: {}".format(median_value))

# Display the DataFrame after mean imputation
df
```

# Imbalanced Data

Imbalanced data is a common issue encountered in data exploration, where one class has significantly fewer instances than the other. This imbalance can adversely affect machine learning models, leading to biased predictions as the model tends to favor the majority class while neglecting the minority class.

## The Problem of Imbalanced Data

Imbalanced data arises when classes in a dataset are not evenly represented. In binary classification, this imbalance occurs when one class has notably more instances than the other. For instance, in a medical dataset predicting a rare disease, the number of patients with the disease may be significantly lower than healthy patients.

This imbalance can result in biased models and inaccurate predictions. Overrepresentation of the majority class may cause the model to favor it, leading to low accuracy, sensitivity, and high false-negative rates for the minority class.



```python
df["Survived"].value_counts()
```

## Methods to Deal with Imbalanced Data

### 1. Oversampling the Minority Class

Oversampling involves increasing instances of the minority class to make it more representative. Techniques like random oversampling, SMOTE (Synthetic Minority Over-sampling Technique), or ADASYN (Adaptive Synthetic Sampling) can be employed.

- **Random oversampling:** Duplicates instances from the minority class randomly. While simple, it may lead to overfitting and reduced generalization.

- **SMOTE:** Generates synthetic instances by interpolating between existing minority class instances. Effective when minority instances are surrounded by majority instances, less effective with sparse minority classes.

- **ADASYN:** Similar to SMOTE but focuses on generating instances that are harder to classify.

```python
df.drop(['Name', "Sex", "Ticket", "Embarked", "PassengerId"], axis = 1, inplace=True)
```

```python
from imblearn.over_sampling import SMOTE
import pandas as pd

X = df.drop('Survived', axis=1) # feature matrix
y = df['Survived'] # target variable


smote = SMOTE()
X_resampled, y_resampled = smote.fit_resample(X, y)
```

```python
y.value_counts()
```

```python
y_resampled.value_counts()
```

### 2. Undersampling the Majority Class

Undersampling involves reducing instances of the majority class to balance the dataset. Techniques include random undersampling, Tomek links, or edited nearest neighbors.

- **Random undersampling:** Randomly removes instances from the majority class. Simple but may lead to information loss and reduced generalization.

- **Tomek links:** Identifies instances close to the decision boundary and removes majority class instances. Effective when majority instances are surrounded by minority instances.

- **Edited nearest neighbors:** Removes misclassified majority class instances. More effective than Tomek links with sparse minority classes and complex decision boundaries.


```python
from imblearn.under_sampling import EditedNearestNeighbours
import pandas as pd

enn = EditedNearestNeighbours()
X_resampled, y_resampled = enn.fit_resample(X, y)
```

```python
y.value_counts()
```

```python
y_resampled.value_counts()
```

### 3. Combination of Oversampling and Undersampling

A combination of oversampling and undersampling aims to balance the dataset, addressing issues of overfitting and information loss. Methods like SMOTE and Tomek links can be used simultaneously.


```python
from imblearn.combine import SMOTETomek
import pandas as pd

smt = SMOTETomek()
X_resampled, y_resampled = smt.fit_resample(X, y)
```

```python
y_resampled.value_counts()
```

## Advantages and Disadvantages of Imbalanced Data Methods

Each method has its pros and cons. Oversampling can effectively increase minority class representation but may lead to overfitting. Undersampling can reduce majority class dominance but may cause information loss. A combination of both may be more effective but can also lead to overfitting. The implementation of these methods using Python and pandas can be facilitated through the imbalanced-learn library, offering functions for oversampling, undersampling, and their combination.

# Outliers

Outliers are data points that significantly deviate from the rest of the observations in a dataset. They can result from measurement errors, data entry mistakes, or rare events. Identifying and managing outliers is crucial in data analysis to prevent skewed results and erroneous conclusions.

## Reasons for Outliers

- **Measurement Error:** Errors in the measurement process can lead to extreme values that do not accurately represent the observed phenomenon.

- **Data Entry Errors:** Mistakes in entering data can introduce values significantly different from other observations.

- **Natural Variation:** Outliers may represent real but rare events that deviate from the typical pattern in the data.

- **Extreme Values:** Outliers can arise from extreme values inherent in the underlying distribution, especially in heavy-tailed distributions.

Regardless of the cause, outliers can distort data exploration and analysis, impacting the reliability of results.

## Methods of Identifying Outliers

1. **Visual Inspection:** Utilizing visualization techniques such as boxplots, scatterplots, and histograms allows the identification of outliers as observations deviating significantly from the overall pattern.

   *Example:* In a scatterplot, an outlier can be visually spotted as a data point deviating markedly from the others.

2. **Descriptive Statistics:** Mean, median, and standard deviation can be employed to identify outliers. Observations falling outside a certain range or threshold are flagged as outliers.

   *Example:* Identifying outliers by calculating the mean and standard deviation, then flagging observations beyond two standard deviations from the mean.

3. **Machine Learning Models:** Algorithms like isolation forests, one-class SVMs, and local outlier factor (LOF) can be applied to identify observations that deviate significantly from the majority.

   *Example:* Using the LOF algorithm to identify outliers in a dataset and flagging observations with a negative outlier factor below a threshold.


```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random data
np.random.seed(123)
data = np.random.normal(0, 1, 100)

# Add an outlier
data[20] = 10

# Create a scatterplot
plt.scatter(range(len(data)), data)
plt.show()
```

```python
import numpy as np

# Generate random data
np.random.seed(123)
data = np.random.normal(0, 1, 100)

data[0] = 10
data[1] = 9

# Calculate mean and standard deviation
mean = np.mean(data)
std = np.std(data)

# Identify outliers
# outliers = [x for x in data if (x < mean - 2 * std) or (x > mean + 2 * std)]
# Mean  2 * Standard Deviation
# The interval Mean  2 * Standard Deviation is used to define a range around the mean where the majority of the data points are expected to lie:

outliers = []
for x in data:
  if (x < mean - 2 * std) or (x > mean + 2 * std) :
    outliers.append(x)

print(outliers)
```

```python
from sklearn.neighbors import LocalOutlierFactor
import numpy as np

# Generate random data
np.random.seed(123)
data = np.random.normal(0, 1, 100)

# Add an outlier
data[0] = 200

# Fit LOF model
lof = LocalOutlierFactor(n_neighbors=5)
lof.fit(data.reshape(-1, 1))

# Identify outliers
outliers = np.where(lof.negative_outlier_factor_ < -2)[0]
print(data[outliers])

```

## Methods of Dealing with Outliers

1. **Removing Outliers:** The simplest approach is to exclude outliers from the dataset. However, caution is needed as this may result in information loss and biased results.

   *Example:* Removing an outlier from a dataset and visualizing the remaining data points in a scatterplot.

2. **Transforming the Data:** Applying mathematical transformations, such as log transformation, helps handle extreme values in the data.

   *Example:* Applying log transformation to a dataset with extreme values and visualizing the transformed data using a histogram.
