# 5. Common BeautifulSoup Commands and Operations


Here are some common commands you can use with BeautifulSoup to extract information from the page.

## 1.Find a specific tag:
You can find the first occurrence of an HTML tag using `soup.find()`.




````python
# Find the first <title> tag
title_tag = soup.find("title")
print(title_tag.text)  # Get the text inside the <title> tag
````

## 2. Find all occurrences of a tag:


````python
# Find all <a> (anchor) tags in the page
all_links = soup.find_all("a")

# Print each link's text and href attribute
for link in all_links:
    print(link.text, "->", link.get('href'))


````

## 3. Extract text from tags:
To extract only the text from an HTML element, use `text`.





````python
# Get the text inside a <p> tag
paragraph = soup.find("p")
print(paragraph.text)
````

````python
all_sections = soup.find_all("section")

section = all_sections[0]

all_divs = section.find_all("div")

print("Number of divs: {}".format(len(all_divs)))

for div in all_divs:
  print(div)
````

**Web Crawling:**

Just like a robot exploring the entire internet, web crawlers, also called spiders or bots, are programs that automatically browse the web and index content for search engines. Their purpose is to visit web pages, collect information, and follow links to discover new pagesbuilding a vast index similar to a giant map of the web. Google, for instance, uses crawlers to update its search index.

**Web Scraping:**

On the other hand, web scraping is a more targeted operation. Rather than visiting the entire web, it focuses on specific sites, extracting precise data like a person picking certain pages from a book. The goal is to gather structured data from websites, such as product prices, news articles, or any other specific content.

Loc and Iloc :

loc and iloc are two important methods in the Pandas library for selecting data from a DataFrame. They are used to access a group of rows and columns by label or integer location. Let's explore each method with examples:

````python
l = [1, 2, 3]
l[0]
````

### Loc

````python
import pandas as pd

# Create a sample DataFrame
data = {'Name': ['Alice', 'Bob', 'Charlie', 'David'],
        'Age': [25, 30, 35, 40],
        'City': ['New York', 'San Francisco', 'Los Angeles', 'Chicago']}

df = pd.DataFrame(data, index=['A', 'B', 'C', 'D'])

df
````

````python
df.iloc[0:4,0:3]
````

````python
# Select a specific row by label
selected_row = df.loc['B']
selected_row
````

````python
# Select a specific element by row and column labels
element = df.loc['B', 'Age']
print(element)
````

````python
# df.loc[[row], [col]]
````

````python
# Select a subset of rows and columns by labels
subset = df.loc[['A', 'B'], ['Age', 'City']]
subset
````

### Iloc

````python
# Select a specific row by integer position
selected_row = df.iloc[2]
selected_row
````

````python
# Select a specific element by row and column positions
element = df.iloc[2, 1]
print(element)
````

````python
# Select a subset of rows and columns by integer positions
subset = df.iloc[0:4, [0, 2]]
subset
````
